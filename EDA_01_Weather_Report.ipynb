# IMPORT LIBRARIES

```python
import numpy 
import pandas 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas_profiling as pp
import sweetviz as sv
%matplotlib inline
```

# LOAD DATASET

```python
data_weather = pd.read_csv('data_clean.csv')
```

```python
data_weather.head(10)
```

```python
data_weather.tail(10)
```

### REMOVING UNWANTED COLUMN Unnamed:0

```python
data_weather.drop(['Unnamed: 0'],axis =1, inplace = True)
```

```python
data_weather.head(10)
```

```python
data_weather.tail(10)
```

### RENAMING COLUMN Solar.R AS Solar R 

```python
data_weather.rename({'Solar.R':'Solar R'},axis =1, inplace=True)
```

```python
data_weather.head(10)
```

```python
data_weather.tail(10)
```

### DUPLICATION HANDLING IN DATASET

```python
data_weather.drop('Temp C',axis=1,inplace=True)
```

```python
data_weather.head()
```

### DATA TYPE CONVERSION

```python
data_weather['Month'] = pd.to_numeric(data_weather['Month'],errors='coerce')
data_weather['Temp'] = pd.to_numeric(data_weather['Temp'],errors='coerce')
data_weather['Weather'] = data_weather['Weather'].astype('category')
```

```python
data_weather.info()
```

### OUTLIER DETECTION

### MISSING VALUES AND IMPUTATIONS

```python
cols = data_weather.columns
colors = ['#000099','#ffff00']
sns.heatmap(data_weather[cols].isnull(),cmap = sns.color_palette(colors))
```

```python
 data_weather[data_weather.isnull().any(axis=1)].head()
```

```python
data_weather.isnull().sum()
```

### FOR CONTINUOUS DATA

#### MEAN IMPUTATION

```python
mean = data_weather['Ozone'].mean()
```

```python
mean
```

#### FILLING MEAN VALUE AT NAN PLACE 

```python
data_weather['Ozone'] = data_weather['Ozone'].fillna(mean)
```

```python
data_weather
```

### FOR CATEGORICAL DATA

#### MODE IMPUTATION

```python
mode = data_weather['Weather'].mode()
```

```python
mode
```

#### FILLING MODE VALUE

```python
data_weather['Weather'] = data_weather['Weather'].fillna(mode)
```

```python
data_weather
```

```python
data_weather.isnull().sum()
```
